3. Write a program in which use all the operators we can use in Python

```python
# Type your code here


a = 10
b = 3


print("a + b =", a + b)   
print("a - b =", a - b) 
print("a * b =", a * b)   
print("a ** b =", a ** b)


c = 5 
print("c =", c)
c += a; print("c += a ->", c)
c -= a; print("c -= a ->", c)
c *= a; print("c *= a ->", c)


print("a == b ->", a == b)
print("a > b  ->", a > b)
print("a < b  ->", a < b)
print("a >= b ->", a >= b)
print("a <= b ->", a <= b)


x, y = True, False
print("x and y ->", x and y)
print("x or y  ->", x or y)
```
